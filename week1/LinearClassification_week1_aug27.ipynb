```python
%pylab inline
import torch
```

```python
x = torch.rand(1000,2)
scatter(*x.numpy().T )
axis('equal')
```

```python

x_in_circle = (x**2).sum(1)<1

def show_1(y):
  scatter(*x.numpy().T, c=y.numpy() )
  axis('equal')
```

```python


show_1(x_in_circle)
```

```python

weights = torch.as_tensor([-1,-1], dtype = torch.float) # 1,1
bias = torch.as_tensor(1, dtype = torch.float) #-1

def classify(x,weights, bias):
  logit = (x * weights[None,:]).sum(1) + bias
  return logit>0

def accuracy1(pred_y):
  return (pred_y ==x_in_circle).float().mean()

pred_y = classify(x,weights, bias)
show_1(pred_y)
print('accuracy=', accuracy1(pred_y))


```

```python
def predict_1(x,weights, bias):
  logit = (x * weights[None,:]).sum(1) + bias
  return 1/(1+(-logit).exp())

def loss_1(p_y):
  return -(x_in_circle.float() * (p_y+1e-10).log() +
  (1-x_in_circle.float()) * (1-p_y+1e-10).log()).mean()

p_y= predict_1(x,weights, bias)
print('loss =',loss_1(p_y))
```

```python
weights = torch.as_tensor([-1,-1], dtype = torch.float) # 1,1
bias = torch.as_tensor(1.5, dtype = torch.float) #-1




p_y= predict_1(x,weights, bias)
pred_y = classify(x,weights, bias)
show_1(pred_y)
print('loss =',loss_1(p_y),'accuracy=', accuracy1(pred_y))
```
